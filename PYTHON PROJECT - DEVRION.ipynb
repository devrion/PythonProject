```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# read CSV into dataframe
df1 = pd.read_csv("aqidaily2009.csv")

pd.to_datetime(df1['Date'])

df1['year'] = df1['Date'].str[-2:]

# create the sqlite DB
db = sqlite3.connect("MyDemo.db")

# convert the pandas dataframe to a sqlite table
df1.to_sql("expenses", db, if_exists="replace")

# view dataframe
df1
```

```python
def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
query = """
SELECT Date, [Overall AQI Value]
FROM expenses;
"""
run_query(query)
```

```python
# Spring                 # Summer                      # Fall                            # Winter
# 21 March to 20 June    # 21 June to 20 September     # 21 September to 20 December     # 21 December to 20 March
```

```python
query = """
SELECT AVG([Overall AQI Value]) AS average, Year
FROM expenses WHERE [Overall AQI Value] > 0 GROUP BY year;
"""


df3 = run_query(query)
```

```python
df3
```

```python
import csv
import matplotlib as plt 
import matplotlib.pyplot as pyplot


df3Values = df3['average'].values
df3toValues = df3['year'].values


bar = pyplot.bar(df3toValues, df3Values, facecolor='red', alpha=0.75)
pyplot.title('Yearly Averages', fontsize=12)
pyplot.xlabel('Year', fontsize=10)
pyplot.ylabel('Air Quality Level', fontsize=10)
bar
```
